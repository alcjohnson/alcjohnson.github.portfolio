```python
import pandas as pd 

file_path = 'invistico_Airline[1].csv'

data = pd.read_csv(file_path)

print(data.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the visual style of the plots
sns.set(style="whitegrid")

# Customer Satisfaction by Travel Class
plt.figure(figsize=(10, 6))
sns.countplot(x='Class', hue='satisfaction', data=data)
plt.title('Customer Satisfaction by Travel Class')
plt.xlabel('Travel Class')
plt.ylabel('Number of Customers')
plt.legend(title='Satisfaction')
plt.show()
```

```python
# Impact of Flight Distance on Satisfaction
plt.figure(figsize=(10, 6))
sns.boxplot(x='satisfaction', y='Flight Distance', data=data)
plt.title('Impact of Flight Distance on Satisfaction')
plt.xlabel('Satisfaction')
plt.ylabel('Flight Distance')
plt.show()
```

```python
# Service Ratings
service_columns = ['Seat comfort', 'Food and drink', 'Gate location', 'Inflight wifi service',
                   'Inflight entertainment', 'Online support', 'Ease of Online booking', 'On-board service',
                   'Leg room service', 'Baggage handling', 'Checkin service', 'Cleanliness', 'Online boarding']

service_data = data[service_columns].mean()

plt.figure(figsize=(12, 8))
service_data.plot(kind='bar')
plt.title('Average Service Ratings')
plt.xlabel('Service')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
# Customer Type and Satisfaction
plt.figure(figsize=(10, 6))
sns.countplot(x='Customer Type', hue='satisfaction', data=data)
plt.title('Customer Type and Satisfaction')
plt.xlabel('Customer Type')
plt.ylabel('Number of Customers')
plt.legend(title='Satisfaction')
plt.show()
```
