```python
import pandas as pd 
```

```python
df = pd.read_csv('world-population.csv')
print(df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
x = df['Year']
y = df['Population']

plt.figure(figsize=(8, 6))
plt.plot(x, y, marker='o', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Line Chart')
plt.grid(True)
plt.show()
```

```python
x = df['Year']
y = df['Population']

plt.figure(figsize=(8, 6))
plt.plot(x, y, drawstyle='steps', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Step Chart')
plt.grid(True)
plt.show()
```
