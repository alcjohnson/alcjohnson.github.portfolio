```python
import matplotlib.pyplot as plt
import pandas as pd

data = {'Year': [2010, 2011, 2012, 2013],
        'Winners': [54, 62, 68, 74]}
df = pd.DataFrame(data)
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize=(8, 6))
plt.bar(df['Year'], df['Winners'], color='pink')
plt.title('Hotdog Contest Winners by Year')
plt.xlabel('Year')
plt.ylabel('Number of Hotdogs Eaten')
plt.show()
```

```python
df['Second Place'] = [49, 60, 65, 72]

plt.figure(figsize=(8, 6))
plt.bar(df['Year'], df['Winners'], label='Winners', color='pink')
plt.bar(df['Year'], df['Second Place'], bottom=df['Winners'], label='Second Place', color='orange')
plt.title('Hotdog Contest Winners vs. Second Place by Year')
plt.xlabel('Year')
plt.ylabel('Number of Hotdogs Eaten')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(df['Winners'], labels=df['Year'], autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Wins by Year')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(df['Winners'], labels=df['Year'], autopct='%1.1f%%', startangle=140)
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Donut Chart: Proportion of Wins by Year')
plt.show()
```
